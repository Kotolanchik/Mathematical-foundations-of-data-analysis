```python
import numpy as np
import matplotlib.pyplot as plt


def func(x, y):
    return np.sqrt(1 - x ** 2 - y ** 2)


x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

Z = func(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def func(x, y):
    return x ** 2 + y ** 2


x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

Z = func(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
import numpy as np
from scipy.optimize import least_squares


def func(params, x, y, z):
    a, b, c = params
    return a * x ** 2 + b * y ** 2


x_points = np.random.uniform(-1, 1, size=100)
y_points = np.random.uniform(-1, 1, size=100)
z_points = np.sqrt(1 - x_points ** 2 - y_points ** 2)

# Аппроксимируем функцию методом Оврагова
params_ovragov = np.random.uniform(-1, 1, size=3)
result_ovragov = least_squares(func, params_ovragov, args=(x_points, y_points, z_points), method='lm')

# Аппроксимируем функцию методом Маркварда
params_markvardt = np.random.uniform(-1, 1, size=3)
result_markvardt = least_squares(func, params_markvardt, args=(x_points, y_points, z_points), method='trf')

# Вычисляем среднюю квадратичную ошибку для каждого метода
mse_ovragov = np.mean((result_ovragov.fun) ** 2)
mse_markvardt = np.mean((result_markvardt.fun) ** 2)

print(f"Средняя квадратичная ошибка метода Оврагова: {mse_ovragov}\n"
      f"Параметры: {result_ovragov.x}\n"
      f"Ошибка: {result_ovragov.cost}\n")
print(f"Средняя квадратичная ошибка метода Маркварда: {mse_markvardt}\n"
      f"Параметры: {result_markvardt.x}\n"
      f"Ошибка: {result_markvardt.cost}\n")
```
