```python
# Лабораторная работа #2.3

import matplotlib.pyplot as plt
import numpy as np


def check_IQR(data):
    IQR = np.percentile(data, 0.75) - np.percentile(data, 0.25)
    upper_limit = np.percentile(data, 0.75) + (IQR * 1.5)
    upper_limit_extreme = np.percentile(data, 0.75) + (IQR * 3)
    print(upper_limit, upper_limit_extreme)
    lower_limit = np.percentile(data, 0.25) - (IQR * 1.5)
    lower_limit_extreme = np.percentile(data, 0.25) - (IQR * 3)
    print(lower_limit, lower_limit_extreme)


x = np.random.randint(1, 1000, 25)
y = np.random.randint(1, 1000, 25)

check_IQR(x)
check_IQR(y)

'''Построение графика'''
plt.title('Отсутствие связи')
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.close()
```

```python
# Лабораторная работа #3

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

m = [328, 454, 172, 625, 609, 546, 202, 736, 485, 766, 4, 313, 328, 344, 36, 453, 563, 343, 375, 28, 312, 361, 297, 437,
     328, 328, 328, 297, 159, 328, 361, 703, 500, 344, 329, 312, 328, 547, 314, 328, 439, 359, 126, 408, 686, 346, 328,
     392, 453, 359]

samples = np.array(m)
mean = np.mean(samples)
var = np.var(samples)
std = np.sqrt(var)

x = np.linspace(min(samples), max(samples), 12)
y_pdf = stats.norm.pdf(x, mean, std)
y_skew_pdf = stats.skewnorm.pdf(x, *stats.skewnorm.fit(samples))
l1, = plt.plot(x, y_pdf, label='Нормальное распределение')
l2, = plt.plot(x, y_skew_pdf, label='Асимметричное нормальное распределение')
n, bins, patches = plt.hist(samples, 12, density=True, facecolor='g', edgecolor='black', alpha=0.75)

plt.xlabel('Значение величины')
plt.ylabel('Частота')
plt.legend((l1, l2), (l1.get_label(), l2.get_label()), loc='upper right')
plt.axis([np.min(m), np.max(m), 0, 0.01])
plt.show()

# H1: Распределение является нормальным
chi_square_test_statistic, p_value = stats.normaltest(samples)
print('Хи-квадрат тест (нормальное распределение):\n'
      f'Тестовая статистика Хи: {chi_square_test_statistic}\n'
      f'p-значение: {p_value}\n')

# H2: Распределение является асимметричным нормальным
chi_square_test_statistic, p_value = stats.skewtest(samples)
print('Хи-квадрат тест (асимметричное нормальное распределение):\n'
      f'Тестовая статистика Хи: {chi_square_test_statistic}\n'
      f'p-значение: {p_value}\n')

# H3: Распределение является экспоненциальным
chi_square_test_statistic, p_value = stats.kstest(samples, 'expon')  # Колмогоров-Смирнов
print('Хи-квадрат тест (экспоненциальное распределение):\n'
      f'Тестовая статистика Хи: {chi_square_test_statistic}\n'
      f'p-значение: {p_value}')
```
