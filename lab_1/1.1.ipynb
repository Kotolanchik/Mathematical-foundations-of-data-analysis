G = 3(A - B)^T * (2C - D), где 
A = [[-2, 3, -4],
    [-3, 5, -2]], 
     
B = [[0, -3, -1],
    [2, 0, -4]], 
     
C = [[0, 4, 1, -3],
    [2, 1, -1, 5]], 
     
D = [[3, 0, 2, -2],
    [0, 3, -3, 0]]

```python
import numpy as np

A = np.array([[-2, 3, -4],
              [-3, 5, -2]])

B = np.array([[0, -3, -1],
              [2, 0, -4]])

C = np.array([[0, 4, 1, -3],
              [2, 1, -1, 5]])

D = np.array([[3, 0, 2, -2],
              [0, 3, -3, 0]])

print(f"A + B = \n{A + B}\n")  # Сложение матриц

print(f"3(A + B) = \n{3 * (A + B)}\n")  # Умножение матрицы на скалярное значение

print(f"3(A + B)^T = \n{3 * (A + B).transpose()}\n")  # Транспонирование и вычитание матриц

print(f"2C - D =\n {2 * C - D}\n")

print(f"G = 3(A - B)^T * (2C - D) =\n {np.dot(3 * (A + B).transpose(), (2 * C - D))}")  # умножение матриц

# 1) Для умножения матриц необходимо обеспечить согласованность: 
# Матрицы А размерности M(строк) x N(столбцов) может умножаться на матрицу размерности N(строк) x P(столбцов
# 
# 2) A * B != B * A (Не коммутативно)
# 
# 3) Пример умножения матриц:
# 
# X = | 100  200  150 |
#     | 50   300  120 |
#     | 80   180  220 |
# 
# Y = | 2.5 |
#     | 3.0 |
#     | 2.2 |
# 
# Z = | 100*2.5 + 200*3.0 + 150*2.2|
#     | 50*2.5 + 300*3.0 + 120*2.2 |
#     | 80*2.5 + 180*3.0 + 220*2.2 |
```

# Задание 32.
Вычислить значение многочлена f(x) от матрицы А.
f(x) = x^3 - x^2 - 1, где
A = [[-2 1],
     [1 0]]

```python
import numpy as np
from numpy import linalg as ln


def f(x):
    return ln.matrix_power(x, 3) - ln.matrix_power(x, 2) - 1


A = np.array([[-2, 1], [1, 0]])
print(f(A))

# Модуль linalg из библиотеки num py, который позволяет осуществлять операции из линейной алгебры для матриц. 

# Пример возведения в степень матрицы:
# | 1  2 | * | 1  2 | = | 1*1 + 2*3  1*2 + 2*4 | = | 7  10 |
# | 3  4 |   | 3  4 |   | 3*1 + 4*3  3*2 + 4*4 |   | 15 22 |
# 
# | 7  10 | * | 1  2 | = | 7*1  + 10*3  7*2 + 10*4  | = | 37  54 |
# | 15 22 |   | 3  4 |   | 15*1 + 22*3  15*2 + 22*4 |   | 81 118 |
```

# Задание 40
Вычислить определитель
| 2 -3 4 3 |
| 4 -2 3 2 |
| 5 -5 2 3 |
| 3 -4 4 3 |

```python
import numpy as np
from numpy import linalg as ln

A = np.array([[2, -3, 4, 3], [4, -2, 3, 2], [5, -5, 2, 3], [3, -4, 4, 3]])

Det_A = ln.det(A)

print(Det_A)

# Свойства:
# * |А| == 0 - матрица является сингулярной, и у нее нет обратной матрицы
# * |А| != 0 - матрица называется невырожденной, у нее есть обратная матрица
# * Если матрица умножается на скаляр, то определитель умножается на это число
# * |A^T| = |A|
# * ЕЯСли две строки(или столбца) линейно независимы, то |A| = 0.   (Лин. завис. - одна строка может быть выражена 
# как линейная комбинация другой строки |[[2 3][4 6]]| = 0)

# Если определитель не равен нулю, то система имеет единственное решение, если равен, то система имеет бесконечно много решений или их нет

# В геометрическом смысле определитель указывает насколько матрица изменят объем или площадь. Если он положителен, 
# то матрица сохраняет ориентацию, и объем или площадь увелич., если отриц., матрица изменяет ориент.
```

# Задание 49
Найти матрицу обратную к данной
А = 
[[6,4,1],
 [5,6,2],
 [1,1,-1]]


```python
import numpy as np
from numpy import linalg as ln

A = np.array([[6, 4, 1], [5, 6, 2], [1, 1, -1]])
A_inv = ln.inv(A)

print(A_inv)

# * Для нахождения обратной матрицы важно убедиться, что она является квадратной
# * Убедиться, что определитель не равне нулю
# 
# Матрица миноров - это матрица, составленная из определителей всех миноров (подматриц),
# полученных из исходной матрицы путем удаления одной или более строк и столбцов.

# Пример:
# A = [[a, b, c],
#      [d, e, f],
#      [g, h, i]]

# M1 = [[e, f], 
#       [h, i]]
# M2 = [[d, f],
#       [g, i]]
# M3 = [[d, e]
#       [g, h]] и тд
# 
# Матрица миноров будет иметь такой вид:
# Minors(A) = [[det(M1), det(M2), det(M3)],
#             [det(M4), det(M5), det(M6)],
#             [det(M7), det(M8), det(M9)]], где det - определитель
# 
# 
# Для нахождения алгебраических дополнений нужно использовать формулу: Aij = (-1)^i+j * Mij, где i - строка, j - столбец
# Далее произвести транспонирование матрицы алгебраических дополнений Adj(A)
# Последним действием будет разделить каждый элемент транспонированной матрицы алг. дополнений на определитель. А^(-1) = (1/det(A))^Adj(A)
```
